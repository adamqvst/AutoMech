```python
import numpy as np
from scipy.fft import rfft, rfftfreq
from scipy import signal
from scipy.signal import find_peaks
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
def analyze_engine_sound(filepath, num_cylinders, dfft_size = 2**16, time_int = [0, 0], frq_int = [0, 1000], num_peaks = 4, ztime = 0.5):
    
    sampling_rate = 0
    samples = []
    sound_clip_length = 0 #length in seconds
    
    #read file
    try:
        sampling_rate, samples = wavfile.read(filepath)
        sound_clip_length = round(len(samples) / sampling_rate, 2)
        print(f'read : {len(samples)} samples from {filepath}')
        print(f'total clip length: {sound_clip_length}s')
    except FileNotFoundError as e:
        print(f'failed to read {filepath}')
        return
        
    #check for valid time interval
    if (time_int[1] > sound_clip_length or time_int[0] == time_int[1]):
        time_int[1] = sound_clip_length
    if (time_int[1] - time_int[0] < ztime):
        ztime = time_int[1] - time_int[0]
        
        
    #disregard second channel
    samples_1D = np.empty(len(samples)) 
    for x in range(0, len(samples)):
        samples_1D[x] = samples[x][0]
        
    #visualize waveform
    fig, (axis0, axis1, axis2) = plt.subplots(3, figsize=(10, 6))
    fig.suptitle(filepath)
    
    step_size = sound_clip_length / len(samples)
    x_axis0 = np.arange(0, sound_clip_length, step_size)              # whole sound clip
    x_axis1 = np.arange(time_int[0], time_int[1], step_size)          # specific interval
    x_axis2 = np.arange(time_int[0], time_int[0] + ztime, step_size)  # zoomed in portion of interval

    axis2.set_xlabel("seconds")
    
    #transform from time space to sample space
    left_time_boundary = int(time_int[0] * sampling_rate)
    right_time_boundary = left_time_boundary + len(x_axis1)    
    right_timez_boundary = left_time_boundary + len(x_axis2)
    
    axis0.plot(x_axis0, samples_1D, linewidth=0.2)
    axis1.plot(x_axis1, samples_1D[left_time_boundary : right_time_boundary], linewidth=0.2)
    axis2.plot(x_axis2, samples_1D[left_time_boundary : right_timez_boundary], linewidth=0.2)
    
    #perform dfft to find frequencies of interest 
    n = dfft_size
    k = np.arange(n)
    T = n / sampling_rate
    xf = k / T
    xf = xf[range(int(n / 2))] # Single sided frequency range
    yf = rfft(samples_1D[0:n-1])
    
    fig_fft = plt.figure(figsize=(20, 8))
    
    #transform from frequency space to sample space
    left_frq_boundary = int(frq_int[0] * len(yf) / (frq_int[1] - frq_int[0]))
    right_frq_boundary = int(frq_int[1] * len(yf) / (frq_int[1] - frq_int[0]))
    
    #find all frequency peaks
    peaks = find_peaks(np.abs(yf[left_frq_boundary : right_frq_boundary]), height = 1, threshold = 1, distance = 10)
    heights = peaks[1]['peak_heights']
    
    #filter out n number of the highest peaks
    x_ind = np.argpartition(heights, -num_peaks)[-num_peaks:]
    print("peaks", len(x_ind))
    y = heights[x_ind]
                                            
    #visualize dfft and annotate peaks with their corresponding frequency and rpm
    for x in x_ind:
        frequency = np.round(peaks[0][x] * (sampling_rate / n), 2)
        rpm = frequency * 60 / (num_cylinders / 2)
        annotation = f'{frequency} Hz ({int(rpm)} rpm)'
        
        plt.annotate(annotation, (frequency + 10, heights[x]))
        
    plt.scatter(peaks[0][x_ind] * (sampling_rate / n), y, marker='x', color='red')
    plt.plot(xf, np.abs(yf), linewidth=0.5)
    plt.xlim(frq_int)
    plt.xlabel("Hz")
    plt.suptitle(f'dfft of {filepath}')
    plt.show()
```

```python
def analyze_engine_sound_ptp(filepath, num_cylinders, lp_freq=100, time_int=[0, 0], ztime = 0.5):
    
    sampling_rate = 0
    samples = []
    sound_clip_length = 0 #length in seconds
    
    #read file
    try:
        sampling_rate, samples = wavfile.read(filepath)
        sound_clip_length = round(len(samples) / sampling_rate, 2)
        print(f'read: {len(samples)} samples from {filepath}')
        print(f'total clip length: {sound_clip_length}s')
    except FileNotFoundError as e:
        print(f'failed to read {filepath}')
    
    #check for valid time interval
    if (time_int[1] > sound_clip_length or time_int[0] == time_int[1]):
        time_int[1] = sound_clip_length
    if (time_int[1] - time_int[0] < ztime):
        ztime = time_int[1] - time_int[0]
        
    #disregard second channel
    samples_1D = np.empty(len(samples)) 
    for x in range(0, len(samples)):
        samples_1D[x] = samples[x][0]
        
    #visualize signals
    fig, (axis0, axis1, axis2) = plt.subplots(3, figsize=(10, 6))
    fig.suptitle(filepath)
    
    step_size = sound_clip_length / len(samples)
    x_axis0 = np.arange(0, sound_clip_length, step_size)              # whole sound clip
    x_axis1 = np.arange(time_int[0], time_int[1], step_size)          # specific interval
    x_axis2 = np.arange(time_int[0], time_int[0] + ztime, step_size)    # first 0.5 seconds of specific interval
    x_axis3 = np.arange(time_int[0], time_int[0] + ztime, step_size)

    axis2.set_xlabel("seconds")
    
    #transform from time space to sample space
    left_time_boundary = int(time_int[0] * sampling_rate)
    right_time_boundary = left_time_boundary + len(x_axis1)    
    right_timez_boundary = left_time_boundary + len(x_axis2)
    
    axis0.plot(x_axis0, samples_1D, linewidth=0.2)
    axis1.plot(x_axis1, samples_1D[left_time_boundary : right_time_boundary], linewidth=0.2)
    axis2.plot(x_axis2, samples_1D[left_time_boundary : right_timez_boundary], linewidth=0.2)
    
    #apply butterworth filter
    sos = signal.butter(10, lp_freq, 'lp', fs=sampling_rate, output='sos')
    filtered_samples_1D = signal.sosfilt(sos, samples_1D)
    
    #normalize filtered signal
    #max = np.max(filtered_samples_1D)
    #min = np.min(filtered_samples_1D)
    #normalized_filtered_samples_1D = (filtered_samples_1D - min) / (max - min)
    
    #polarize data
    polarized_filtered_samples_1D = np.empty(len(filtered_samples_1D)) 
    for x in range(0, len(filtered_samples_1D)):
        polarized_filtered_samples_1D[x] = (1.0 if (filtered_samples_1D[x] >= 0.0) else -1.0)
    
    #calculate distance in samples between peaks
    peak_distance = sampling_rate / (10000 / 60)
    
    #count number of peaks between time interval
    peaks = find_peaks(polarized_filtered_samples_1D[left_time_boundary : right_time_boundary])
    
    #visualize filtered signals
    filtered_fig, (f_axis0, f_axis1, f_axis2, f_axis3) = plt.subplots(4, figsize=(10,8))
    filtered_fig.suptitle('filtered signals')
    
    f_axis0.plot(x_axis0, filtered_samples_1D, linewidth=0.2, color='red')
    f_axis1.plot(x_axis1, filtered_samples_1D[left_time_boundary : right_time_boundary], linewidth=0.2, color='red')
    f_axis2.plot(x_axis2, filtered_samples_1D[left_time_boundary : right_timez_boundary], linewidth=0.2, color='red')
    f_axis3.plot(x_axis2, polarized_filtered_samples_1D[left_time_boundary : right_timez_boundary], linewidth=0.2, color='red')
    f_axis3.scatter((peaks[0] + left_time_boundary) / sampling_rate, np.ones(len(peaks[0])), marker='x', color='green')
    f_axis3.set_xlim([time_int[0], time_int[0] + 0.5])
    
    number_of_peaks = len(peaks[0])
    number_of_samples = len(samples_1D[left_time_boundary : right_time_boundary])
    interval_clip_length = number_of_samples / sampling_rate
    peaks_per_second = number_of_peaks / (time_int[1] - time_int[0])
    calculated_rpm = peaks_per_second * 60 / (num_cylinders / 2)
    
    print(f'peak distance: {peak_distance}')
    print(f'number of peaks: {number_of_peaks}')
    print(f'number of samples: {number_of_samples}')
    print(f'clip length: {np.round(interval_clip_length, 2)}s')
    print(f'peaks per second: {np.round(peaks_per_second, 2)}')
    print(f'calculatede rpm: {int(calculated_rpm)}rpm')
```
